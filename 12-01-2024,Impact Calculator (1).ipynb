```python
# waste generated is taken as average of carbo,sulphur and  enegy content percentage
import pandas as pd
def calculate_carbon_footprint():
    # Input values
    electricity_consumption = float(input("Enter annual electricity consumption in kWh: "))
    transportation_distance = float(input("Enter annual transportation distance in miles: "))
    # Load waste data from CSV
    waste_data = pd.read_csv("waste.csv")
    # Extract relevant columns from waste data
    sulphur_percentage = waste_data["Sulphur_Percentage"]
    carbon_percentage = waste_data["Carbon_Percentage"]
    energy_content = waste_data["Energy_Content"]
    # Use the average values
    average_sulphur_percentage = sulphur_percentage.mean()
    average_carbon_percentage = carbon_percentage.mean()
    average_energy_content = energy_content.mean()
    # Conversion factors
    electricity_emission_factor = 0.4  # Example emission factor for electricity in kgCO2/kWh
    transportation_emission_factor = 0.2  # Example emission factor for transportation in kgCO2/mile
    # Adjusted based on average values from waste data
    waste_emission_factor = (average_sulphur_percentage + average_carbon_percentage + average_energy_content) / 3 
    # Calculate carbon footprint
    electricity_emission = electricity_consumption * electricity_emission_factor
    transportation_emission = transportation_distance * transportation_emission_factor
    waste_emission = waste_emission_factor 
    total_emission = electricity_emission + transportation_emission + waste_emission
    print(f"Your annual carbon footprint is {total_emission:.2f} kgCO2.")
if __name__ == "__main__":
    calculate_carbon_footprint()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the CSV file into a DataFrame
df = pd.read_csv('waste.csv')
# Calculate the average for each column
average_values = df.mean()
# Plotting the bar graph with exact values
plt.figure(figsize=(14, 8))
plt.subplot(2, 3, 5)
sns.barplot(x=df.columns, y=average_values)
plt.title('Environmental Impact Metrics Distribution')
# Display the exact values on the plot
for i, value in enumerate(average_values):
    plt.text(i, value, f'{value:.2f}', ha='center', va='bottom', color='red')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Load the CSV file
df = pd.read_csv('waste.csv')
# Convert columns to numeric if they are not already
df['Carbon_Percentage'] = pd.to_numeric(df['Carbon_Percentage'], errors='coerce')
df['Moisture_Percentage'] = pd.to_numeric(df['Moisture_Percentage'], errors='coerce')
# Visualize Carbon Percentage vs. Moisture Percentage
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Carbon_Percentage', y='Moisture_Percentage', data=df)
plt.title('Carbon Percentage vs. Moisture Percentage')
plt.xlabel('Carbon Percentage')
plt.ylabel('Moisture Percentage')
plt.show()
```
